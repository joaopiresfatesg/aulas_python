Peça a idade da pessoa. Mostre:
▶ Criança (0–12)
▶ Adolescente (13–17)
▶ Adulto (18 ou mais)
(Use if, elif e else).

```python
idade=int(input("digite sua idade"))
print(idade)
if(idade<13):
    print("criança")
elif(idade > 18):
    print("adulto")
else:
    print ("adolescente")
```
